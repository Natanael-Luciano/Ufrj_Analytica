```python
text = open('Texto.txt','rt', encoding = 'utf-8')
lines = text.readlines()
usable_text = ''
for i in lines:
    usable_text += i.split('\n')[0] + ' '
usable_text = usable_text.lower() 
for car in [':', ',',';','.','?','!']:
    x = usable_text.split(car)
    usable_text = ''.join([i + ' ' for i in x] )
word_count = {}
for word in usable_text.split(' '):
    try:
        word_count[word] += 1
    except:
        word_count[word] = 1

print('Entrada:')
text = open('Texto.txt','rt', encoding = 'utf-8')
print(text.read())

print('\nSAÍDA')
for i in word_count.items():
    print(i)
```
